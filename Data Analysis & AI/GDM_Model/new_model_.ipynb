# **Feature engineering**

```python
df_model=data.copy()
```

```python
print("original:", df_model.shape)
```

```python
cols_to_drop = [
    'id',
    'gravidity',
    'parity',
    'gestational_age_weeks',
    'hemoglobin_gdl',
    'anemia_status',
    'hiv_status',
    'anc_visits',
    'delivery_mode',
    'pregnancy_outcome',
    'risk_level'
]

df_model = df_model.drop(columns=cols_to_drop, errors='ignore')
```

```python
print("original:", df.shape)
```

```python
# convert bmi into categories
import pandas as pd

bins = [0, 18.5, 24.9, 29.9, float('inf')]
labels = ['Underweight', 'Normal', 'Overweight', 'Obese']

# new column
df_model['bmi_category'] = pd.cut(df_model['bmi_pre_pregnancy'], bins=bins, labels=labels)

#encoding
df_model['bmi_category_encoded'] = df_model['bmi_category'].cat.codes

print(df_model[['bmi_pre_pregnancy', 'bmi_category', 'bmi_category_encoded']].head())
```

```python
# drop Useless bmi's columns
df_model = df_model.drop(['bmi_pre_pregnancy', 'bmi_category'], axis=1)

print(df_model.columns)
```

```python
print("after bmi:", df_model.shape)
```

```python
# apply Mean Arterial Pressure (MAP) equation
# equation: (Systolic + 2*Diastolic) / 3
df_model['MAP'] = (df_model['systolic_bp_mmhg'] + 2 * df_model['diastolic_bp_mmhg']) / 3

# Rounding the results
df_model['MAP'] = df_model['MAP'].round(2)


print(df_model[['systolic_bp_mmhg', 'diastolic_bp_mmhg', 'MAP']].head())
```

```python
# drop Useless
df_model = df_model.drop(['systolic_bp_mmhg', 'diastolic_bp_mmhg'], axis=1)

print(df_model.columns)
```

```python
# drop rare values
values_to_drop = ['hemorrhage', 'severe_anemia', 'eclampsia']

df_model = df_model[~df_model['primary_complication'].isin(values_to_drop)]

#print remaining values
print("Remaining values")
print(df_model['primary_complication'].value_counts())

#Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_model['target_encoded'] = le.fit_transform(df_model['primary_complication'])


mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(mapping)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score, confusion_matrix, classification_report
from imblearn.over_sampling import SMOTE

# --------------------------
# 1️⃣ Prepare data

# Multi-label targets
df_model['gestational_diabetes'] = (df_model['primary_complication'] == 'gestational_diabetes').astype(int)


# Features (remove target)
X = df_model[['age_years','fasting_glucose_mgdl','proteinuria','bmi_category_encoded','MAP']]
y_diabetes = df_model['gestational_diabetes']


# --------------------------
# 2️⃣ Preprocessing
numeric_features = ['age_years','fasting_glucose_mgdl','proteinuria','MAP']
categorical_features = ['bmi_category_encoded']

preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numeric_features),
    ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_features)
])

# --------------------------
# 3️⃣ Train/test split
X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(X, y_diabetes, test_size=0.2, random_state=42, stratify=y_diabetes)

# --------------------------
# 4️⃣ Handle imbalance with SMOTE
smote = SMOTE(random_state=42)
X_train_d_res, y_train_d_res = smote.fit_resample(X_train_d, y_train_d)


# --------------------------
# 5️⃣ Define models
models = {
    'LogisticRegression': LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42),
    'RandomForest': RandomForestClassifier(n_estimators=200, class_weight='balanced', random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss',
                             scale_pos_weight=(y_train_d==0).sum()/(y_train_d==1).sum(),
                             random_state=42)
}

# --------------------------
# 6️⃣ Function to train and evaluate
def train_eval(model, X_train, y_train, X_test, y_test, label, threshold=0.5):
    pipe = Pipeline([
        ('preprocess', preprocessor),
        ('model', model)
    ])
    pipe.fit(X_train, y_train)
    probs = pipe.predict_proba(X_test)[:,1]
    preds = (probs >= threshold).astype(int)

    print(f"===== {label} - {model.__class__.__name__} =====")
    print("Accuracy:", accuracy_score(y_test, preds))
    print("F1 Score:", f1_score(y_test, preds))
    print("Precision:", precision_score(y_test, preds))
    print("Recall:", recall_score(y_test, preds))
    print("ROC AUC:", roc_auc_score(y_test, probs))
    print("Confusion Matrix:\n", confusion_matrix(y_test, preds))
    print("Classification Report:\n", classification_report(y_test, preds))
    print("\n")

    return pipe, probs

# --------------------------
# 7️⃣ Train and evaluate models for Gestational Diabetes
diabetes_pipes = {}
diabetes_probs = {}
threshold_diabetes = 0.7  # adjustable threshold to reduce False Positives

for name, model in models.items():
    pipe, probs = train_eval(model, X_train_d_res, y_train_d_res, X_test_d, y_test_d, "Gestational Diabetes", threshold=threshold_diabetes)
    diabetes_pipes[name] = pipe
    diabetes_probs[name] = probs

# --------------------------
# 9️⃣ Probabilities for the full dataset (optional)
diabetes_probs_full = diabetes_pipes['RandomForest'].predict_proba(X)[:,1]  # Example: using best model


results = pd.DataFrame({
    'diabetes_prob': diabetes_probs_full,

})

print(results.head())
```
